```python
import requests
from IPython.core.display import display_html, HTML
import pandas as pd
import pprint


def pd_print(obj):
  """输出HTML"""
  if obj is None:
    return
  display_html(HTML(pd.json_normalize(obj).to_html(justify='left')))


url = 'https://peps.python.org/api/peps.json'
headers = {
    'Content-type': 'application/json'
}
params = {
}
data = {
}

print(url)
print(headers)
print(params)

r = requests.get(url=url, headers=headers, params=params, verify=False)
data = r.json()
pprint.pprint(data['1'])

peps = []
for _, v in data.items():
  peps.append({
    'number': v['number'],
    'title': v['title'],
    'type': v['type'],
    'status': v['status'],
    # 'url': v['url']
  })
pd_print(peps)
```
